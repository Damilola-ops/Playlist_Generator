```python
artist_id=['3tVQdUvClmAT7URs9V3rsp','0Y3agQaa6g2r0YmHPOO9rh','4qYpTEJThZ8FC8KzyFrSWW',
          '46pWGuE3dSwY3bMMXGBvVS','3wcj11K77LjEY1PkEazffa','1XavfPKBpNjkOfxHINlMHF','5yOvAmpIR7hVxiS6Ls5DPO',
          '687cZJR45JO7jhk1LHIbgq','379IT6Szv0zgnw4xrdu4mu','4Ns55iOSe1Im2WU2e1Eym0']
artists=['Wizkid','Davido','Boj','Rema','Burna Boy','Joeboy','Omah lay','Tems','Ladipoe','Simi']
```

```python
dictt={}
for i in artists:
    for j in artist_id:
        dictt[i]=j
        artist_id.remove(j)
        break
```

```python
dictt
```

```python

idd=[]
album=[]
artist_list=[]
song_name=[]
image=[]
url=[]
for i in dictt.items():
    for k in sp.artist_top_tracks(i[1])['tracks']:
        idd.append(i[1])
        album.append(k['album']['name'])
        artist_list.append(i[0])
        song_name.append(k['name'])
        url.append(k['external_urls']['spotify'])
        for l in k['album']['images']:
            image.append(l['url'])
```
