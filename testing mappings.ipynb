```python
import pandas as pd
import numpy as np
import re
```

```python
df=pd.read_csv('spotify_transformed.csv')
```

```python
df.drop('Unnamed: 0',axis='columns',inplace=True)
```

```python
df['for_lyrics']
```

```python
df['for_lyrics']=df['for_lyrics'].apply(lambda x:x.split('\n')[1:])
```

```python
df['for_lyrics']=df['for_lyrics'].apply(lambda x:str(x))
```

```python
import string
string.punctuation
```

```python
df['for_lyrics']=df['for_lyrics'].apply(lambda x:re.sub(r"""[+[\]-]""",' ',x))
```

```python
dictt=dict(zip(df.song_name,df.for_lyrics))
```

```python
new_dict={}
for i,j in dictt.items():
    for k in j.split(','):
        new_dict[str(k).strip()]=i
```
